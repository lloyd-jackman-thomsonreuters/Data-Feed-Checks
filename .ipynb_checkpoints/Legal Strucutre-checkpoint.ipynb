```python
import pandas as pd
import numpy as np
from IPython.display import display
from fuzzywuzzy import fuzz
```

```python
gfd1 = pd.read_csv("...\\GFD-1\\LegalStructure.txt", sep="\t", encoding='cp1252')
gfd1.head()
```

```python
cfg = pd.read_csv("...\\CFG\\LegalStructure.txt", sep="\t")
cfg.head()
```

```python
cfg.equals(gfd1)
```

```python
df = cfg.merge(gfd1, how='outer', on="LegalStructCode", suffixes=("_CFG", "_GFD-1"))
check = []
for n in range(len(df.iloc[:,1])):
    if df.iloc[:,1][n] == df.iloc[:,2][n]:
        check.append("Match")
    elif str(df.iloc[:,1][n]) == "nan":
        check.append("Retired")
    elif str(df.iloc[:,2][n]) == "nan":
        check.append("Introduced in new system only")
    else:
        check.append("Updated field")
df["Check"] = np.array(check)
df.to_excel("Legal Structure.xlsx", index=False)
df.sample(n=25)
```
